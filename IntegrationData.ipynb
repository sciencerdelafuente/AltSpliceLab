```python
import pandas as pd
import networkx as nx
import csv
from xml.dom import minidom
import xml.etree.ElementTree as ET
from xml.etree.ElementTree import tostring
import numpy as np # linear algebra
```

```python
#Read Files of Genome Variables and Metadata
metadata = pd.read_csv('Metadata.csv', header=0, sep='\t', low_memory=False)
metadata_assembly = pd.read_csv('Metadata_Assembly.csv', header=0, sep='\t', low_memory=False)
genome_variables = pd.read_csv('Output_Genome_Variables.csv', header=0, sep='\t', low_memory=False)
refseq_eukaryotes = pd.read_csv('Refseq_Eukaryotes.csv', header=0, sep='\t', low_memory=False)

#Merge all variables into a single File called 'Genome_Variables.csv'

#Assembly accession name
list_assembly_accession = []
list_organism_name = []
list_asm_name = []
list_genome_size = []
list_total_gene_count = []
list_protein_coding_gene_count = []
list_non_coding_gene_count = []
list_Taxonomy = []
list_genome_size_ungapped = []
list_scaffold_count = []
list_contig_count = []


#We first add the variables of interst from "Refseq_Eukaryotes.csv" to "genome_variables"
it=list(range(0, len(genome_variables)))   
for i in it:
    posf=len(genome_variables.FileName[i])
    substring = genome_variables.FileName[i][4:posf]
    pos=substring.find('_')+4
    assembly_accession = genome_variables.FileName[i][0:pos]
    list_assembly_accession.append(assembly_accession)
    row=refseq_eukaryotes.loc[refseq_eukaryotes['assembly_accession'] == assembly_accession]
    
    organism_name=row.organism_name.tolist()[0]
    list_organism_name.append(organism_name)
    
    asm_name=row.asm_name.tolist()[0]
    list_asm_name.append(asm_name)
    
    genome_size=row.genome_size.tolist()[0]
    list_genome_size.append(genome_size)
    
    total_gene_count=row.total_gene_count.tolist()[0]
    list_total_gene_count.append(total_gene_count)
    
    protein_coding_gene_count=row.protein_coding_gene_count.tolist()[0]
    list_protein_coding_gene_count.append(protein_coding_gene_count)
    
    non_coding_gene_count=row.non_coding_gene_count.tolist()[0]
    list_non_coding_gene_count.append(non_coding_gene_count)
    
    Taxonomy=row.Taxonomy.tolist()[0]
    list_Taxonomy.append(Taxonomy)
    
    genome_size_ungapped=row.genome_size_ungapped.tolist()[0]
    list_genome_size_ungapped.append(genome_size_ungapped)
    
    scaffold_count=row.scaffold_count.tolist()[0]
    list_scaffold_count.append(scaffold_count)
    
    contig_count=row.contig_count.tolist()[0]
    list_contig_count.append(contig_count)
       
    

        
genome_variables["assembly_accession"]=list_assembly_accession
genome_variables["organism_name"]=list_organism_name
genome_variables["asm_name"]=list_asm_name
genome_variables["total_gene_count"]=list_total_gene_count
genome_variables["protein_coding_gene_count"]=list_protein_coding_gene_count
genome_variables["non_coding_gene_count"]=list_non_coding_gene_count
genome_variables["Taxonomy"]=list_Taxonomy
genome_variables["genome_size_ungapped"]=list_genome_size_ungapped
genome_variables["scaffold_count"]=list_scaffold_count
genome_variables["contig_count"]=list_contig_count



#We now add all data from "Metadata Assembly" to "genome_variables"---------------

list_spanned_gaps = []
list_unspanned_gaps = []
list_scaffold_count_assembly = []
list_scaffold_N50 = []
list_contig_count_assembly = []
list_contig_N50 = []
list_total_gap_length = []
list_gc_perc = []

metadata_assembly = metadata_assembly.replace(r'\n','', regex=True) 
it=list(range(0, len(genome_variables)))   
for i in it:
    accession=genome_variables.assembly_accession[i]
    row=metadata_assembly.loc[metadata_assembly['refseq_assembly_accession'] == accession]
    
    spanned_gaps=row.spanned_gaps.tolist()[0]
    list_spanned_gaps.append(spanned_gaps)
    
    unspanned_gaps=row.unspanned_gaps.tolist()[0]
    list_unspanned_gaps.append(unspanned_gaps)
    
    scaffold_count=row.scaffold_count.tolist()[0]
    list_scaffold_count_assembly.append(scaffold_count)
    
    scaffold_N50=row.scaffold_N50.tolist()[0]
    list_scaffold_N50.append(scaffold_N50)
 
    contig_count=row.contig_count.tolist()[0]
    list_contig_count_assembly.append(contig_count)
    
    contig_N50=row.contig_N50.tolist()[0]
    list_contig_N50.append(contig_N50)
    
    total_gap_length=row.total_gap_length.tolist()[0]
    list_total_gap_length.append(total_gap_length)
     
    gc_perc=row.gc_perc.tolist()[0]
    list_gc_perc.append(gc_perc)


genome_variables["spanned_gaps"]=list_spanned_gaps
genome_variables["unspanned_gaps"]=list_unspanned_gaps
genome_variables["scaffold_count_assembly"]=list_scaffold_count_assembly
genome_variables["scaffold_N50"]=list_scaffold_N50
genome_variables["contig_count_assembly"]=list_contig_count_assembly
genome_variables["contig_N50"]=list_contig_N50
genome_variables["total_gap_length"]=list_total_gap_length
genome_variables["gc_perc"]=list_gc_perc




#We finally add all data from "Metadata" to "genome_variables"---------------

metadata = metadata.replace(r'\n','', regex=True) 


list_Genes_total = []
list_Genes_protein_coding = []
list_mRNAs = []
list_CDSs = []
list_CDSs_model_refseq = []
list_CDSs_known_refseq = []
list_CDSs_fully_supported = []
list_CDSs_ab_initio = []
list_Length_count_genes = [] 
list_Length_mean_genes = []
list_Length_count_transcripts = []
list_Length_mean_transcripts = []
list_Length_count_cds = []
list_Length_mean_cds = []
list_Length_count_exons = []
list_Length_mean_exons = []
list_Length_count_introns = []
list_Length_mean_introns = []
list_Num_proteins = []
list_Num_transcripts = []
list_Num_proteins_hsapiens = []
list_Num_transcripts_hsapiens = []
list_Num_rnaseq_reads = []
list_Num_tissues = []
list_Num_runs = []

list_remove= []
list_remove_accession = []

it=list(range(0, len(genome_variables)))   
for i in it:
    accession=genome_variables.assembly_accession[i]
    row=metadata.loc[metadata['Accession'] == accession]
    
    inv=1
    if(len(row.Accession.tolist())==0):
        list_remove_accession.append(i) 
        
        accession=genome_variables.organism_name[i]
        rowc=metadata.loc[metadata['ScientificName'] == accession]
        if(len(rowc.ScientificName.tolist())==0):
            list_remove.append(i)
            inv=0            
        else:
            row=rowc

            
        
        
        
          
    if(len(row.Accession.tolist())==0):
        Genes_total = 0
    else:
        Genes_total=row.Genes_total.tolist()[0]
    list_Genes_total.append(Genes_total)
    

    if(len(row.Accession.tolist())==0):
        Genes_protein_coding = 0
    else:
        Genes_protein_coding=row.Genes_protein_coding.tolist()[0]
    list_Genes_protein_coding.append(Genes_protein_coding)
    
    
    if(len(row.Accession.tolist())==0):
        mRNAs = 0
    else:
        mRNAs=row.mRNAs.tolist()[0]
    list_mRNAs.append(mRNAs)
    

    if(len(row.Accession.tolist())==0):
        CDSs = 0
    else:
        CDSs=row.CDSs.tolist()[0]
    list_CDSs.append(CDSs)
    
    
    if(len(row.Accession.tolist())==0):
        CDSs_model_refseq = 0
    else:
        CDSs_model_refseq=row.CDSs_model_refseq.tolist()[0]
    list_CDSs_model_refseq.append(CDSs_model_refseq)
    
 
    if(len(row.Accession.tolist())==0):
        CDSs_known_refseq = 0
    else:
        CDSs_known_refseq=row.CDSs_known_refseq.tolist()[0]
    list_CDSs_known_refseq.append(CDSs_known_refseq)
    
 
    if(len(row.Accession.tolist())==0):
        CDSs_fully_supported = 0
    else:
        CDSs_fully_supported=row.CDSs_fully_supported.tolist()[0]
    list_CDSs_fully_supported.append(CDSs_fully_supported)
    
 
   
    if(len(row.Accession.tolist())==0):
        CDSs_ab_initio = 0
    else:
        CDSs_ab_initio=row.CDSs_ab_initio.tolist()[0]
    list_CDSs_ab_initio.append(CDSs_ab_initio)
    
    
    
    if(len(row.Accession.tolist())==0):
        Length_count_genes = 0
    else:
        Length_count_genes=row.Length_count_genes.tolist()[0]
    list_Length_count_genes.append(Length_count_genes)
    
 
   
    
    if(len(row.Accession.tolist())==0):
        Length_mean_genes = 0
    else:
        Length_mean_genes=row.Length_mean_genes.tolist()[0]
    list_Length_mean_genes.append(Length_mean_genes)
    
    
    
    if(len(row.Accession.tolist())==0):
        Length_count_transcripts = 0
    else:
        Length_count_transcripts=row.Length_count_transcripts.tolist()[0]
    list_Length_count_transcripts.append(Length_count_transcripts)
     
   
    
    if(len(row.Accession.tolist())==0):
        Length_mean_transcripts = 0
    else:
        Length_mean_transcripts=row.Length_mean_transcripts.tolist()[0]
    list_Length_mean_transcripts.append(Length_mean_transcripts)
    
 
    
    if(len(row.Accession.tolist())==0):
        Length_count_cds = 0
    else:
        Length_count_cds=row.Length_count_cds.tolist()[0]
    list_Length_count_cds.append(Length_count_cds)
    
 
    
    if(len(row.Accession.tolist())==0):
        Length_mean_cds = 0
    else:
        Length_mean_cds=row.Length_mean_cds.tolist()[0]
    list_Length_mean_cds.append(Length_mean_cds)

   
    
    if(len(row.Accession.tolist())==0):
        Length_count_exons = 0
    else:
        Length_count_exons=row.Length_count_exons.tolist()[0]
    list_Length_count_exons.append(Length_count_exons)
    
 

    if(len(row.Accession.tolist())==0):
        Length_mean_exons = 0
    else:
        Length_mean_exons=row.Length_mean_exons.tolist()[0]
    list_Length_mean_exons.append(Length_mean_exons)
    
 

    if(len(row.Accession.tolist())==0):
        Length_count_introns = 0
    else:
        Length_count_introns=row.Length_count_introns.tolist()[0]
    list_Length_count_introns.append(Length_count_introns)
    
 

    if(len(row.Accession.tolist())==0):
        Length_mean_introns = 0
    else:
        Length_mean_introns=row.Length_mean_introns.tolist()[0]
    list_Length_mean_introns.append(Length_mean_introns)
    
 

    if(len(row.Accession.tolist())==0):
        Num_proteins = 0
    else:
        Num_proteins=row.Num_proteins.tolist()[0]
    list_Num_proteins.append(Num_proteins)
    
 

    if(len(row.Accession.tolist())==0):
        Num_transcripts = 0
    else:
        Num_transcripts=row.Num_transcripts.tolist()[0]
    list_Num_transcripts.append(Num_transcripts)
    
 

    if(len(row.Accession.tolist())==0):
        Num_proteins_hsapiens = 0
    else:
        Num_proteins_hsapiens=row.Num_proteins_hsapiens.tolist()[0]
    list_Num_proteins_hsapiens.append(Num_proteins_hsapiens)
    
 
    if(len(row.Accession.tolist())==0):
        Num_transcripts_hsapiens = 0
    else:
        Num_transcripts_hsapiens=row.Num_transcripts_hsapiens.tolist()[0]
    list_Num_transcripts_hsapiens.append(Num_transcripts_hsapiens)
    
        
    if(len(row.Accession.tolist())==0):
        Num_rnaseq_reads = 0
    else:
        Num_rnaseq_reads=row.Num_rnaseq_reads.tolist()[0]
    list_Num_rnaseq_reads.append(Num_rnaseq_reads)
    
 

    if(len(row.Accession.tolist())==0):
        Num_tissues = 0
    else:
        Num_tissues=row.Num_tissues.tolist()[0]
    list_Num_tissues.append(Num_tissues)
    
 
    if(len(row.Accession.tolist())==0):
        Num_runs = 0
    else:
        Num_runs=row.Num_runs.tolist()[0]
    list_Num_runs.append(Num_runs)
    
         
       

genome_variables["Genes_total"]=list_Genes_total
genome_variables["Genes_protein_coding"]=list_Genes_protein_coding
genome_variables["mRNAs"]=list_mRNAs
genome_variables["CDSs"]=list_CDSs
genome_variables["CDSs_model_refseq"]=list_CDSs_model_refseq
genome_variables["CDSs_known_refseq"]=list_CDSs_known_refseq
genome_variables["CDSs_fully_supported"]=list_CDSs_fully_supported
genome_variables["CDSs_ab_initio"]=list_CDSs_ab_initio
genome_variables["Length_count_genes"]=list_Length_count_genes
genome_variables["Length_mean_genes"]=list_Length_mean_genes
genome_variables["Length_count_transcripts"]=list_Length_count_transcripts
genome_variables["Length_mean_transcripts"]=list_Length_mean_transcripts
genome_variables["Length_count_cds"]=list_Length_count_cds
genome_variables["Length_mean_cds"]=list_Length_mean_cds
genome_variables["Length_count_exons"]=list_Length_count_exons
genome_variables["Length_mean_exons"]=list_Length_mean_exons
genome_variables["Length_count_introns"]=list_Length_count_introns
genome_variables["Length_mean_introns"]=list_Length_mean_introns
genome_variables["Num_proteins"]=list_Num_proteins
genome_variables["Num_transcripts"]=list_Num_transcripts
genome_variables["Num_proteins_hsapiens"]=list_Num_proteins_hsapiens
genome_variables["Num_transcripts_hsapiens"]=list_Num_transcripts_hsapiens
genome_variables["Num_rnaseq_reads"]=list_Num_rnaseq_reads
genome_variables["Num_tissues"]=list_Num_tissues
genome_variables["Num_runs"]=list_Num_runs


genome_variables.drop(list_remove, inplace=True)
genome_variables.to_csv('Genome_Variables.csv', sep='\t')
genome_variables[0:10]



        
                

        
        


```
